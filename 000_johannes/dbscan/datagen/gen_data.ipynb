```python
import random
import math
import matplotlib.pyplot as plt
import os
```

```python
outDir = os.path.join('.', 'output')
os.makedirs(outDir, exist_ok = True)
```

```python
npoints = 10000
centers = [ (1,2), (7,2), (10,5) ]
sigmas = [ (1.3, 0.6), (0.75, 0.75), (2.5, .5) ]
```

```python
ns = [ math.floor(npoints / len(centers)) ] * len(centers)
ns[-1] += sum(ns) - npoints
```

```python
pointsLists = []
for i in range(0, len(centers)):
    points = []
    n = ns[i]
    center = centers[i]
    sigma = sigmas[i]
    for j in range(0, n):
        points.append( (
            center[0] + random.gauss(mu = 0, sigma = sigma[0]),
            center[1] + random.gauss(mu = 0, sigma = sigma[1])
        ) )
    pointsLists.append(points)
    
```

```python
plt.figure(figsize=(4,3))
for points in pointsLists:
    plt.scatter([p[0] for p in points], [p[1] for p in points], s=2)
plt.show()
```

```python
pointsShuffled = []
for points in pointsLists:
    pointsShuffled.extend(points)
random.shuffle(pointsShuffled)
```
